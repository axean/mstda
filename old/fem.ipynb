```python
import json
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tqdm.notebook import tqdm
from functools import partial
import gudhi as gd
import scipy.spatial as spatial
from more_itertools import powerset
```

```python
def compute_centroid(points):
    return np.sum(points, axis = 0) / points.shape[0]

points = np.random.rand(10,2)
tri = spatial.Delaunay(points)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
ax.triplot(points[:,0], points[:,1], tri.simplices)

for i,p in enumerate(points):
    ax.text(*tuple(p+0.01),str(i))

centroids = list()
for face in tri.simplices:
    for sf in [sf for sf in powerset(face) if len(sf) > 0]:
        c = compute_centroid(points[sf, :])
        centroids.append(c)
        ax.scatter(x = c[0], y = c[1], marker = "x", color = "red", s = 50)
centroids = np.asarray(centroids)
```
